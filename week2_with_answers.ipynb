## Python Data Structures: Quick Overview

Here’s a quick summary of four fundamental Python data structures you've learned:

- **List**: Ordered and changeable collection, allows duplicates.
- **Tuple**: Ordered and unchangeable (immutable), allows duplicates.
- **Set**: Unordered, unindexed collection of unique elements.
- **Dictionary**: Collection of key-value pairs, used for structured data access.

These are the building blocks for handling and organizing data.


````python
# List: a sequence of values (can repeat)
transactions = [100, 200, 150, 200]
print("List:", transactions)

# Tuple: like a list, but immutable
account_info = ("Alice", "Savings", 3500)
print("Tuple:", account_info)

# Set: unique values only
branches = {"North", "South", "West", "South"}
print("Set:", branches)

# Dictionary: key-value pairs
customer = {
    "name": "Alice",
    "balance": 3500,
    "account_type": "Savings"
}
print("Dictionary:", customer)
````

## Python Operators: Quick Recap

Operators help you perform operations on values and variables.

- **Arithmetic**: `+`, `-`, `*`, `/`, `//`, `%`, `**`
- **Comparison**: `==`, `!=`, `<`, `>`, `<=`, `>=`
- **Logical**: `and`, `or`, `not`
- **Assignment**: `=`, `+=`, `-=`, etc.

You'll use these for calculations, comparing values, and control logic in your projects.


````python
# Arithmetic
a = 10
b = 3
print("Addition:", a + b)
print("Division:", a / b)
print("Modulus:", a % b)

# Comparison
print("Is a equal to b?", a == b)

# Logical
x = True
y = False
print("x and y:", x and y)
print("x or y:", x or y)
````

## Loops in Python

Loops allow you to repeat code multiple times:

- **for loop**: Used to iterate over a sequence (like a list or a range).
- **while loop**: Runs as long as a condition is `True`.


````python
# For loop
print("For loop")
for i in [1, 2, 3]:
    print("Item:", i)

print("\nWhile loop")
# While loop
count = 0
while count < 3:
    print("Count is:", count)
    count += 1
````

## TypeError: unhashable type: 'dict'

You might see this error when trying to use a dictionary inside a `set()` or as a key in another dictionary.

This is because:
- A `set` only accepts **hashable** (immutable) items.
- Dictionaries are **mutable** → they can change → so they are **not hashable**.

### How to Fix It
If you need to store dictionaries in a set or compare them for duplicates:
- Convert each dictionary to a **tuple of values**, or
- Use a unique string representation (like turning it into a sorted tuple of items).


````python
seen = set()
duplicates = []

for tx in transactions:
    tx_tuple = (tx['customer'], tx['type'], tx['amount'], tx['branch'])
    if tx_tuple in seen:
        duplicates.append(tx)
    else:
        seen.add(tx_tuple)

print("Duplicate transactions:", duplicates)
````

## Hint: Formatting with f-Strings?

**f-strings** (formatted string literals) let you insert variables directly into strings using `{}`.

They start with the letter `f` before the string:

```python
name = "Alice"
age = 30
print(f"My name is {name} and I am {age} years old.")


## Hint: Aligning Output with f-Strings

When printing tables or columns, you can use f-strings with alignment to make your output clean and easy to read.

### Alignment Options
Inside the `{}` of an f-string, you can add formatting like this:

```python
print(f"{value:<10}")  # Left-align in a space 10 characters wide
print(f"{value:>10}")  # Right-align in 10 characters
print(f"{value:^10}")  # Center-align in 10 characters


````python
print(f"{'Name':<10} | {'Score':>5}")
print(f"{'Alice':<10} | {95:>5}")
print(f"{'Bob':<10} | {87:>5}")
````

````python
# Step 1: Build valid_transactions
seen = set()
valid_transactions = []

for tx in transactions:
    tx_tuple = (tx["customer"], tx["type"], tx["amount"], tx["branch"])
    if isinstance(tx["amount"], (int, float)) and tx_tuple not in seen:
        valid_transactions.append(tx)
        seen.add(tx_tuple)

# Step 2: Build summary
summary = {}
for tx in valid_transactions:
    name = tx["customer"]
    ttype = tx["type"]
    amt = tx["amount"]
    if name not in summary:
        summary[name] = {"deposit": 0, "withdrawal": 0}
    summary[name][ttype] += amt

# Step 3: Print results
print(f"{'Customer':<8} | {'Deposits':<9} | {'Withdrawals':<12} | {'Net Balance'}")
print("-" * 50)
balances = {}
for name, s in summary.items():
    net = s["deposit"] - s["withdrawal"]
    balances[name] = net
    print(f"{name:<8} | ${s['deposit']:<8} | ${s['withdrawal']:<11} | ${net}")
````
